```python
from romgw.config.env import PROJECT_ROOT
from romgw.plotting.core import plot_same_mode, plot_same_params
from romgw.typing.core import BBHSpinType, DatasetType, ModeType, MODE_VALUES
from romgw.waveform.base import FullWaveform
from romgw.waveform.dataset import FullWaveformDataset
```

```python
bbh_spin: BBHSpinType = "NS"
dataset: DatasetType = "test"
wf_dir = lambda mode: (
    PROJECT_ROOT / "data" / bbh_spin / dataset / mode / "full" / "raw"
)
```

```python
seed = 0

modes = {}
for mode in MODE_VALUES:
    modes[mode] = FullWaveform.from_directory(wf_dir(mode), seed)
plot_same_params(modes)
```

```python
mode: ModeType = "2,2"
n = 16
seed = 0

waveforms = FullWaveformDataset.from_directory(wf_dir(mode), n, seed)
plot_same_mode(waveforms, verbose=False)
```
